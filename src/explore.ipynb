```python
from sklearn.preprocessing import MinMaxScaler
df['job_n'] = pd.factorize(df['job'])[0]
df['marital_n'] = pd.factorize(df['marital'])[0]
df['education_n'] = pd.factorize(df['education'])[0]
df['default_n'] = pd.factorize(df['default'])[0]
df['housing_n'] = pd.factorize(df['housing'])[0]
df['loan_n'] = pd.factorize(df['loan'])[0]
df['contact_n'] = pd.factorize(df['contact'])[0]
df['month_n'] = pd.factorize(df['month'])[0]
df['day_of_week_n'] = pd.factorize(df['day_of_week'])[0]
df['poutcome_n'] = pd.factorize(df['poutcome'])[0]
df['y_n'] = pd.factorize(df['y'])[0]
```

```python
scaler = MinMaxScaler()
scal_df = ['job_n','marital_n','education_n','default_n','housing_n','loan_n','contact_n','month_n','day_of_week_n','poutcome_n','y_n','age',
             'duration','campaign','pdays','previous','emp.var.rate','cons.price.idx','cons.conf.idx','euribor3m','nr.employed']
```

```python
scal_df
```

```python
arr = scaler.fit_transform(df[scal_df])
df_1 = pd.DataFrame(arr, index=df.index, columns=scal_df)
```

```python
df_1
```

```python
from sklearn.model_selection import train_test_split

X = df_1.drop('y_n', axis=1)
y = df_1['y_n']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=4)
```

```python
from sklearn.feature_selection import chi2, SelectKBest
selection_model = SelectKBest(chi2, k=5)
selection_model.fit(X_train, y_train)
var_1 = selection_model.get_support()
```

```python
selection_model
```

```python
var_1
```

```python
X_train_sel = pd.DataFrame(selection_model.transform(X_train), columns=X_train.columns.values[var_1])
X_test_sel = pd.DataFrame(selection_model.transform(X_test), columns=X_test.columns.values[var_1])
```

```python
X_train.columns.values[var_1]
```

```python
X_train_sel.head()
```

```python
X_test_sel.head()
```

```python
X_train_sel['y_n'] = list(y_train)
X_test_sel['y_n'] = list(y_test)
```

```python
train_data = X_train_sel
test_data = X_test_sel
```

```python
train_data
```

```python
X_train = train_data.drop(['y_n'], axis=1)
```

```python
X_train
```

```python
y_train = train_data['y_n']
```

```python
X_test = test_data.drop(['y_n'], axis=1)
```

```python
y_test = test_data['y_n']
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV

# We define the parameters that we want to adjust by hand
hyperparams = {
    "C": [0.001, 0.01, 0.1, 1, 10, 100, 1000],
    "penalty": ["l1", "l2", "elasticnet", None],
    "solver": ["newton-cg", "lbfgs", "liblinear", "sag", "saga"]
}
```

```python
grid = GridSearchCV(logreg, hyperparams, scoring = "accuracy", cv = 5)
grid
```

```python
grid.fit(X_train, y_train)
```

```python
print(f"Best hyperparameters: {grid.best_params_}")
```

```python
model_grid = LogisticRegression(penalty = "l2", C = 0.1, solver = "sag")
model_grid.fit(X_train, y_train)
y_pred = model_grid.predict(X_test)

grid_accuracy = accuracy_score(y_test, y_pred)
grid_accuracy
```
